# Q. 9) Which individual country has the Highest No. of TV Shows ?

```python
netflix_tvshow = netflix_data[netflix_data["Category"]=="TV Show"]
```

```python
netflix_tvshow
```

```python
netflix_tvshow["Country"].value_counts()
```

```python
netflix_tvshow["Country"].value_counts().head(10)
```

```python
# Individual country that has the Highest No. of TV Shows : 
netflix_tvshow["Country"].value_counts().head(1)
```

```python
# Plotting graph for 10 Countries that has the Highest No. of TV Shows : 
Top10countries = netflix_tvshow["Country"].value_counts().head(10)
```

```python
Store = pandas.DataFrame({"TotalNumber_of_Tvshows":Top10countries})
Store
```

```python
plt.figure(figsize=(14,10))
plt.plot(Store,label = "Number of Tv Shows")
plt.xlabel("Country Names",size = 13)
plt.ylabel("Number of Tv Shows",size = 13)
plt.title("TOTAL NUMBER OF TV SHOWS IN EACH COUNTRY",size = 13)
plt.grid(visible = True,which="major")
plt.legend(loc="best",shadow=True,fancybox = True,fontsize="large")
```

# Q. 10) How can we sort the dataset by Year ?

```python
netflix_data["Release_Year"] = netflix_data["Date_N"].dt.year
```

```python
netflix_data
```

```python
# Sorting the movies from latest year(Recent) to oldest year 
netflix_data.sort_values(by="Release_Year",ascending=False)
```

```python
# Sorting the movies from oldest year to latest year
netflix_data.sort_values(by="Release_Year",ascending=True)
```

# Q. 11) Find all the instances where: Category is 'Movie' and Type is 'Dramas' or Category is 'TV Show' & Type is 'Kids' TV'.


```python
netflix_data[(netflix_data["Category"]=="Movie") & (netflix_data["Type"]=="Dramas") | (netflix_data["Category"]=="TV Show") & (netflix_data["Type"]=="Kid's TV")]
```

```python
# Other Method : 
first_condition = (netflix_data["Category"]=="Movie") & (netflix_data["Type"]=="Dramas")
```

```python
first_condition
```

```python
second_condition = (netflix_data["Category"]=="TV Show") & (netflix_data["Type"]=="Kid's TV")
```

```python
second_condition
```

```python
netflix_data[first_condition | second_condition]
```
